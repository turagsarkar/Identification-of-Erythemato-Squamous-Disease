```python
#data cleaning
import numpy as np
from numpy import NaN
new_data=data.replace("?",NaN)
new_data
new_data['Age (linear)'].unique()
```

```python
new_data['Age (linear)'].isnull().sum()
```

```python
new_data[new_data['Age (linear)'].isnull()]
```

```python
new_data['Age (linear)']=new_data['Age (linear)'].astype('Int64')
```

```python
new_data.dtypes
```

```python
new_data['Age (linear)'].fillna(int(new_data['Age (linear)'].mean()),inplace=True)
```

```python
#outlier detection for Numerical feature by boxplot
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(new_data,y='Age (linear)')
plt.figure(figsize=(50,25))
```

```python
#calculate z-score for outlier detection
#add new column z-score
new_data['Z-score']=(new_data['Age (linear)']-new_data['Age (linear)'].mean())/new_data['Age (linear)'].std()
new_data['Z-score'].head()
```
